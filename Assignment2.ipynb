```python
import pandas as pd
```

```python
worker = pd.read_csv('Worker.csv')
payitem = pd.read_csv('PayItem.csv')
job = pd.read_csv('Job.csv')
worker_itemized = pd.read_csv('Worker_Itemized_Amount.csv')
```

```python
worker_itemized_payitem = pd.merge(worker_itemized, payitem, on='PayItemID', how='inner')
worker_job = pd.merge(worker, job, on='WorkCode', how='inner')
final_merge = pd.merge(worker_itemized_payitem, worker_job, on='WorkerID', how='inner')
```

```python
final_merge = final_merge[['WorkerID', 'PayItemID', 'PayItemDescription', 'Status', 'HireDate', 'EndDate']]
```
